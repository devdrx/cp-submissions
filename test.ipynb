```python
import matplotlib.pyplot as plt

# Values of Z and their corresponding probabilities
Z_values = [3, 4, 5, 6, 7]
P_Z = [1/6, 1/6, 1/3, 1/6, 1/6]

# (a) Plot the histogram for the probability distribution
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(Z_values, P_Z, color='skyblue', edgecolor='black')
plt.xlabel('Z')
plt.ylabel('P(Z)')
plt.title('Probability Distribution of Z')

# (b) Compute the cumulative distribution function F(Z)
F_Z = [sum(P_Z[:i+1]) for i in range(len(P_Z))]

# Plot the distribution function
plt.subplot(1, 2, 2)
plt.step(Z_values, F_Z, where='mid', color='red')
plt.xlabel('Z')
plt.ylabel('F(Z)')
plt.title('Distribution Function of Z')
plt.ylim(0, 1.1)

plt.tight_layout()
plt.show()
```
